# Git Push Notebook
This notebook demonstrates how to create another notebook file in the repository, add it to Git, and push the repository to GitHub.

```python
import json
import os
import subprocess
from pathlib import Path

ROOT = Path(r"c:\Users\jigar\OneDrive\Desktop\New folder")
NOTEBOOK_PATH = ROOT / "demo_notebook.ipynb"

print(f"Repository root: {ROOT}")
print(f"Demo notebook path: {NOTEBOOK_PATH}")
```

```python
demo_nb = {
    "cells": [
        {
            "cell_type": "markdown",
            "metadata": {},
            "source": [
                "# Demo Notebook\n",
                "This notebook is created by `git_push.ipynb` and demonstrates a valid Jupyter notebook structure."
            ]
        },
        {
            "cell_type": "code",
            "execution_count": null,
            "metadata": {},
            "outputs": [],
            "source": [
                "print('Hello from the generated notebook!')\n"
            ]
        }
    ],
    "metadata": {
        "kernelspec": {
            "display_name": "Python 3",
            "language": "python",
            "name": "python3"
        },
        "language_info": {
            "name": "python"
        }
    },
    "nbformat": 4,
    "nbformat_minor": 5
}

NOTEBOOK_PATH.write_text(json.dumps(demo_nb, indent=2), encoding='utf-8')
print(f"Saved demo notebook to {NOTEBOOK_PATH}")
```

```python
def run_git(args):
    command = [r"C:\Program Files\Git\cmd\git.exe"] + args
    result = subprocess.run(command, cwd=ROOT, capture_output=True, text=True)
    print('Command:', ' '.join(command))
    print('Return code:', result.returncode)
    print('STDOUT:\n', result.stdout)
    print('STDERR:\n', result.stderr)
    return result

# Track the notebook and the generated demo notebook
run_git(['status', '--short'])
```

```python
# Add the notebook files to git and commit them
run_git(['add', 'git_push.ipynb', 'demo_notebook.ipynb', '.gitignore'])
run_git(['status', '--short'])
run_git(['commit', '-m', 'Add git push helper notebook and .gitignore'])

# Set the remote and push to GitHub
run_git(['remote', 'add', 'origin', 'https://github.com/Jigar-10/smart-plantcare.git'])
run_git(['branch', '-M', 'main'])
run_git(['push', '-u', 'origin', 'main'])
```
